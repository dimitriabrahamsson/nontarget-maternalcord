```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('R01_300_12_pos_CleanFnaBaCShC.csv')
df2 = pd.read_csv('R01_300_12_neg_CleanFnaBaCShC.csv')
```

```python
df1a = df1.loc[:, 'chem_id_s1']
df1b = df1.loc[:, 'BB02375-M331_01':'BB02674-C479_02']
df1c = df1.loc[:, 'BC01006-651C_01':'BC01537-845C_02']
```

```python
df1c = pd.concat([df1c, df1a], axis=1)
df1b = pd.concat([df1b, df1a], axis=1)
df1b = df1b.set_index('chem_id_s1')
df1c = df1c.set_index('chem_id_s1')
```

```python
df1m = df1b.groupby(np.arange(len(df1b.columns))//2, axis=1).mean()
df1b_column_names = df1b.columns.values

df1b_column_names = pd.DataFrame({'sample_names': df1b_column_names})
df1b_column_names.sample_names = df1b_column_names.sample_names.str.replace('_01', '')
df1b_column_names.sample_names = df1b_column_names.sample_names.str.replace('_02', '')
df1b_column_names = df1b_column_names.sample_names.drop_duplicates()
df1b_column_names = df1b_column_names.reset_index(drop=True)
df1m.columns = df1b_column_names
#df1b_column_names.values
```

```python
df1m = df1m.reset_index()
df1m.columns = df1m.columns.str.replace('index', 'chem_id_s1')

df1m.columns.tolist()
df1m_column_names = df1m.columns.values
df1m_column_names = pd.DataFrame({'column names': df1m_column_names})
df1m_column_names['column names'] = df1m_column_names['column names'].str.split('-').apply(lambda x: x[-1])
df1m.columns = df1m_column_names['column names']

df1b = df1m
```

```python
df1m = df1c.groupby(np.arange(len(df1c.columns))//2, axis=1).mean()
df1c_column_names = df1c.columns.values

df1c_column_names = pd.DataFrame({'sample_names': df1c_column_names})
df1c_column_names.sample_names = df1c_column_names.sample_names.str.replace('_01', '')
df1c_column_names.sample_names = df1c_column_names.sample_names.str.replace('_02', '')
df1c_column_names = df1c_column_names.sample_names.drop_duplicates()
df1c_column_names = df1c_column_names.reset_index(drop=True)
df1m.columns = df1c_column_names
#df1c_column_names.values
```

```python
df1m = df1m.reset_index()
df1m.columns = df1m.columns.str.replace('index', 'chem_id_s1')

df1m.columns.tolist()
df1m_column_names = df1m.columns.values
df1m_column_names = pd.DataFrame({'column names': df1m_column_names})
df1m_column_names['column names'] = df1m_column_names['column names'].str.split('-').apply(lambda x: x[-1])
df1m.columns = df1m_column_names['column names']

df1c = df1m
```

```python
df1 = pd.merge(df1b, df1c, on='chem_id_s1', how='outer')
```
