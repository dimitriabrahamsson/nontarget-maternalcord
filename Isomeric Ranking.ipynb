```python
df = df.loc[df['FOUND_BY'] != 'NO_MATCH']
```

```python
df
```

```python
df['DATA_SOURCES'] = df['DATA_SOURCES'].replace('-', '')
df['NUMBER_OF_PUBMED_ARTICLES'] = df['NUMBER_OF_PUBMED_ARTICLES'].replace('-', '')
df['PUBCHEM_DATA_SOURCES'] = df['PUBCHEM_DATA_SOURCES'].replace('-', '')
df['CPDAT_COUNT'] = df['CPDAT_COUNT'].replace('-', '')

#df['DATA_SOURCES'] = df['DATA_SOURCES'].str.extract('(\d+)', expand=False)
df['NUMBER_OF_PUBMED_ARTICLES'] = df['NUMBER_OF_PUBMED_ARTICLES'].str.extract('(\d+)', expand=False)
df['PUBCHEM_DATA_SOURCES'] = df['PUBCHEM_DATA_SOURCES'].str.extract('(\d+)', expand=False)
df['CPDAT_COUNT'] = df['CPDAT_COUNT'].str.extract('(\d+)', expand=False)

df['DATA_SOURCES'] = df['DATA_SOURCES'].astype(float)
df['NUMBER_OF_PUBMED_ARTICLES'] = df['NUMBER_OF_PUBMED_ARTICLES'].astype(float)
df['PUBCHEM_DATA_SOURCES'] = df['PUBCHEM_DATA_SOURCES'].astype(float)
df['CPDAT_COUNT'] = df['CPDAT_COUNT'].astype(float)
```

```python
DF=df

DF=df.reset_index()
```

```python
DF
DF = DF.fillna(1)
```

```python
group=DF.groupby(['INPUT'])
```

```python
def ratio(dataframe):
    isomer_count=dataframe.DTXSID.count()
    ratio=1/isomer_count
    dataframe['Blind_Probability']=ratio
    return(dataframe['Blind_Probability'])

def data(DF):
    data_ratio = DF['DATA_SOURCES']/DF.DATA_SOURCES.sum()
    DF['DATA_RATIO']=data_ratio
    return(DF['DATA_RATIO'])

def pubmed(DF):
    PUBMED_RATIO=DF['NUMBER_OF_PUBMED_ARTICLES'].divide(DF.NUMBER_OF_PUBMED_ARTICLES.sum(),fill_value=0)
    DF['PUBMED_RATIO']=PUBMED_RATIO
    return(DF['PUBMED_RATIO'])

def pubchem(DF):
    PUBCHEM_RATIO=DF['PUBCHEM_DATA_SOURCES'].divide(DF.PUBCHEM_DATA_SOURCES.sum(),fill_value=0)
    DF['PUBCHEM_RATIO']=PUBCHEM_RATIO
    return (DF['PUBCHEM_RATIO'])

def CPDAT(DF):
    CPDAT_RATIO=DF['CPDAT_COUNT'].divide(DF.CPDAT_COUNT.sum(),fill_value=0)
    DF['CPDAT_RATIO']=CPDAT_RATIO
    return (DF['CPDAT_RATIO'])
```

```python
blind=pd.DataFrame(group.apply(ratio))

data_ratio = pd.DataFrame(group.apply(data))

pubmed = pd.DataFrame(group.apply(pubmed))

pubchem = pd.DataFrame(group.apply(pubchem))

CPDAT = pd.DataFrame(group.apply(CPDAT))
```

```python
concat = pd.concat([blind,data_ratio,pubmed,pubchem,CPDAT],axis=1)

#concat1 = concat.droplevel(1)

#concat2=concat1.reset_index()
```
