```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('R01_300_12_pos_Av.csv')
```

```python
df1['Mass_av'] = (df1['Mass_s1'] + df1['Mass_s2'])/2
df1['Retention Time_av'] = (df1['Retention Time_s1'] + df1['Retention Time_s2'])/2
```

```python
df2 = df1.set_index('iso_id')
```

```python
df2 = df2.loc[:, 'M331':'845C']
```

```python
df2 = df2.T
```

```python
df2
```

```python
df2.columns.name = None
```

```python
df2
```

```python
df2_cm = df2.corr()
df2_cm = df2_cm.where(pd.np.triu(pd.np.ones(df2_cm.shape), k=1).astype(bool)).stack().reset_index() #extract upper triangle of the matrix
df2_cm = df2_cm.dropna()
df2_cm = df2_cm.rename(index = str, columns = {'level_0':'rows', 'level_1':'columns', 0:'r-coefficient'})
df2_cm['id'] = df2_cm['rows'] + '+' + df2_cm['columns']
```

```python
# Exract r-coefficients > 0.9
df2_cm = df2_cm.loc[df2_cm['r-coefficient'] >= 0.5]
```

```python
df2_cm 
```

```python
df2_cm['iso_id'] = df2_cm['rows']
```

```python
df1_loc = df1[['iso_id', 'Mass_av', 'Retention Time_av']]
```

```python
df1_loc.columns = df1_loc.columns.str.replace('Mass_av', 'Mass_row')
```

```python
df2row = pd.merge(df2_cm, df1_loc, on='iso_id', how = 'left')
```

```python
df2row
```

```python
df2row['iso_id_2'] = df2row['columns']

df1_loc = df1[['iso_id', 'Mass_av', 'Retention Time_av']]
df1_loc.columns = df1_loc.columns.str.replace('Mass_av', 'Mass_column')
df1_loc.columns = df1_loc.columns.str.replace('iso_id', 'iso_id_2')
```

```python
df2 = pd.merge(df2row, df1_loc, on='iso_id_2', how = 'left', suffixes = ('_r', '_c'))
```

```python
df2 = df2.drop('id', axis=1)
```

```python
df2
```

```python
df2['Mass_diff'] = df2['Mass_row'] - df2['Mass_column']
```

```python
df2['Mass_diff']
```

```python
# Sodium adducts (ESI+ only)
df2['Sodium_Adtest'] = (1-((df2['Mass_diff'] + 22.989769)/22.989769))*10**6
df2['Sodium_columnIsAdduct'] = np.where(((df2['Sodium_Adtest'] >= 0)&(df2['Sodium_Adtest'] <= 15)), 1, 0)
df2['Sodium_rowIsAdduct'] = np.where(((df2['Sodium_Adtest'] <= 0)&(df2['Sodium_Adtest'] >= -15)), 1, 0)
```

```python
df2['Sodium_columnIsAdduct'].sum(), df2['Sodium_rowIsAdduct'].sum()
```

```python
df2
```

```python
# Potasium adducts (ESI+ only)
df2['Potasium_Adtest'] = (1-((df2['Mass_diff'] + 38.963706)/38.963706))*10**6
df2['Potasium_columnIsAdduct'] = np.where(((df2['Potasium_Adtest'] >= 0)&(df2['Potasium_Adtest'] <= 15)), 1, 0)
df2['Potasium_rowIsAdduct'] = np.where(((df2['Potasium_Adtest'] <= 0)&(df2['Potasium_Adtest'] >= -15)), 1, 0)
```

```python
df2['Potasium_columnIsAdduct'].sum(), df2['Potasium_rowIsAdduct'].sum()
```

```python
# Ammonium adducts (ESI+ only)
df2['Ammonium_Adtest'] = (1-((df2['Mass_diff'] + 18.033826)/18.033826))*10**6
df2['Ammonium_columnIsAdduct'] = np.where(((df2['Ammonium_Adtest'] >= 0)&(df2['Ammonium_Adtest'] <= 15)), 1, 0)
df2['Ammonium_rowIsAdduct'] = np.where(((df2['Ammonium_Adtest'] <= 0)&(df2['Ammonium_Adtest'] >= -15)), 1, 0)
```

```python
df2['Ammonium_columnIsAdduct'].sum(), df2['Ammonium_rowIsAdduct'].sum()
```

```python
# ACN adducts (ESI+ only)
df2['ACN_Adtest'] = (1-((df2['Mass_diff'] + 41.026549)/41.026549))*10**6
df2['ACN_columnIsAdduct'] = np.where(((df2['ACN_Adtest'] >= 0)&(df2['ACN_Adtest'] <= 15)), 1, 0)
df2['ACN_rowIsAdduct'] = np.where(((df2['ACN_Adtest'] <= 0)&(df2['ACN_Adtest'] >= -15)), 1, 0)
```
